```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
from datetime import datetime
df = pd.read_csv("data/train.csv")
```

```python
df["dt"] = df["datetime"].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
df["year"] = df["dt"].apply(lambda x: x.year)
df["year_normalized"] = df["year"] - df["year"].min()
df["month"] = df["dt"].apply(lambda x: x.month)
df["month_normalized"] = df["year_normalized"] * 12 + df["month"]
df["day"] = df["dt"].apply(lambda x: x.day)
df["hour"] = df["dt"].apply(lambda x: x.hour)
df["timestamp"] = df["dt"].apply(lambda x: (x - datetime(1970,1,1)).total_seconds())
df["timestamp"] = df["timestamp"] - df["timestamp"].min()
```

```python
sns.boxplot(x="hour", y="count", hue="weather", data=df)
```

```python
sns.boxplot(x="month_normalized", y="count", data=df)
```
